```python
import random

with open("input.txt", "r") as file:
    pop, total_run = map(int, file.readline().strip().split())

    scores, players = {}, []
    for _ in range(pop):
        name, run = file.readline().strip().split()
        scores[name] = int(run)
        players.append(name)

def chromo(players):
    sel_players, sel = [], []
    for _ in range(pop):
        name = random.choice(players)
        sel_players.append(name)

    for i in range(pop):
        sel.append(1 if players[i] in sel_players else 0)
    return sel

def fit(sel):
    value = sum(scores[players[i]] for i in range(pop) if sel[i] == 1)
    return value

def crossover(parents):
    children = []
    for i in range(0, 4, 2):
        cross_idx = random.randint(0, pop-1)
        child_1 = parents[i][:cross_idx] + parents[i+1][cross_idx:]
        child_2 = parents[i+1][:cross_idx] + parents[i][cross_idx:]
        child_1, child_2 = mutate(child_1, child_2)

        children.append(child_1)
        children.append(child_2)
    return children

def mutate(chromo_1, chromo_2):
    chromo_1[random.randint(0, pop-1)] = random.randint(0, 1)
    chromo_2[random.randint(0, pop-1)] = random.randint(0, 1)
    return chromo_1, chromo_2

def run_chase(players):
    def notation(lst):
        return ''.join(map(str, lst))

    parents_list = [chromo(players) for _ in range(4)]
    for _ in range(1000):
        for parent in parents_list:
            if fit(parent) == total_run:
                return f'{players}\n{notation(parent)}'
        else:
            parents_list = crossover(parents_list)

    return f'{players}\n-1'

print(run_chase(players))
```
