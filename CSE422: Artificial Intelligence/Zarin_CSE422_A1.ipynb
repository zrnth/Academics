```python
import fileinput


my_dict = {}
my_list = []
my_parent_dict = {}                                                 #parent_node_storing



for line in fileinput.input(files='input_file.txt'):
    line = line.strip()
    elements = line.split(" ")
    my_list.append([elements[0], elements[1]])

    for i in range(2, len(elements) - 1, 2):                           # Extracting_cur_node
        node = elements[0]

        if node not in my_parent_dict:
            my_parent_dict[node] = []


        if node not in my_dict:
            my_dict[node] = []

        my_dict[node].append([elements[i], elements[i + 1]])
        my_parent_dict[node].append(elements[i])


nodes = list(my_dict.keys())                  # Extracting_node_from_main dict


path = []
visited_nodes = []                             #storing visited places
my_non_min_list = []
total_cost = 0


def pikachu(start, goal):                     # traversing graph and finding my way
    visited_nodes.append(start)
    values = my_dict[start]

    if not values:                            # Checking if there are no destinations (values) for the current node
        return "no dest"

    else:
        min_value = float('inf')
        min_path_node = None


        for v in values:
            if v[0] in visited_nodes:           #checking oi jaygay gese naki jay nai
                pass

            else:
                if int(v[1]) < min_value:
                    min_value = int(v[1])
                    min_path_node = v[0]

        global total_cost                      #cost updating
        total_cost += min_value


        if min_path_node is not None:          #checking rasta paise kina
            path.append(min_path_node)


        else:
            print("NO PATH FOUND")             #rasta na paile
            return


    if path[-1] == goal:                    #goal node checking
        print("Path: ", path)
        print("Total distance: ", total_cost)
    else:
        pikachu(path[-1], goal)

shuru = input("Starting node: ")
sesh = input("Destination: ")

pikachu(shuru, sesh)
```
